## Acquiring Data for World Cases.

### Importing necessary packages

```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

### Getting World data 

```python
url = 'https://www.worldometers.info/coronavirus/'

source = requests.get(url)
print(source)
```

**Response 200 denotes that the connection to the URL is successful.**

```python
soup = BeautifulSoup(source.content,'html.parser')
soup.title
```

Verifyng the title to confirm whether we are parsing the right data

```python
table = soup.find_all('table')
print(len(table))
```

```python
t = table[0]
```

```python
country = []
total_cases = []
new_cases = []
total_deaths = []
new_deaths = []
recovered = []
active_cases = []

rows = t.find_all('tr')[9:-8] #removing irrelevant columns.
#tr - parents class of the table

for row in rows:
    col = row.find_all('td')
    country.append(col[0].text.strip())
    total_cases.append(col[1].text.strip())
    new_cases.append(col[2].text.strip())
    total_deaths.append(col[3].text.strip())
    new_deaths.append(col[4].text.strip())
    recovered.append(col[5].text.strip())
    active_cases.append(col[6].text.strip())
    
print(country)
print(total_cases)
print(new_cases)
print(total_deaths)
print(new_deaths)
print(recovered)
print(active_cases)
```

```python
df = pd.DataFrame(list(zip(country,total_cases,new_cases,total_deaths,new_deaths,recovered,active_cases)),
                 columns=['Country','Total_cases','New_cases','Total_deaths','New_deaths','Recovered','Active_cases'])

df
```

## Data Cleaning

### Replacing empty/null values to NaN format

```python
df.replace(r'^\s*$',np.nan,regex=True, inplace=True)
df
```

```python
df.head(20)
```

**We are replacing NaN values of only two columns to zero. The reason being is, we cannot fill some values based on median/mean values for these columns. It varies day to day.**

```python
df.New_cases = df['New_cases'].replace({np.nan:0})
df.New_deaths = df['New_deaths'].replace({np.nan:0})
```

**Checking NaN values for remaining columns**

```python
df.isna().sum().sort_values
```

**As we checked, two columns of 'Recovered' columns have N/A which is not considered as NaN. So how can we fill this data. Zero? or with some other value?. Lets do a quick math to calculate for the 3rd index row**

Total Cases: 171253
Total Deaths: 26771
Active Cases: 144138

Recovered = Total_cases - (Total_deaths + Active_cases) = 171253 - (26771 + 144138) = 344

```python
df.loc[df['Recovered']=='N/A'] #344 #250
```

```python
df.loc[3]
```

```python
df.loc[3,'Recovered'] = 344
df.loc[12,'Recovered'] = 250
```

```python
df.loc[[3,12]]
```

**We have successfully replaced the values of 'N/A' in 3rd and 12th row.Lets check Whether we have to do similar imputation based on logics in some columns which have NaN**

```python
df.loc[df['Total_deaths'].isna()]
```

```python
df.loc[df['Recovered'].isna()]
```

**As we checked, we haven't found any logics as like earlier. So we can replace NaN values with Zeros**

```python
df.Total_deaths = df['Total_deaths'].replace({np.nan:0})
df.Recovered = df['Recovered'].replace({np.nan:0})
```

```python
df.isna().sum().sort_values(ascending = False)
```

**Now we have imputated Null values, its time for data cleaning '+,' symbols and convert to int datatype. Why Integer? When the data type is in object, you cannot plot different plots and get insights.**

```python
df.head()
```

```python
def plus_comma(num):
    s = str(num).replace('+','').replace(',','')
    s1 = int(s)
    return s1
```

```python
df['New_cases'] = df['New_cases'].map(plus_comma)
```

```python
df['New_deaths'] = df['New_deaths'].map(plus_comma)
df['Total_cases'] = df['Total_cases'].map(plus_comma)
df['Total_deaths'] = df['Total_deaths'].map(plus_comma)
df['Recovered'] = df['Recovered'].map(plus_comma)
df['Active_cases'] = df['Active_cases'].map(plus_comma)
```

**Converted symbols and comma using map function. Now if you check the datatype, all columns turned into integer**

```python
df.head()
```

```python
df.info()
```

### Feature reduction 

**New cases and New Deaths can be added with Active Cases and Total deaths respectively.**

Total Cases = Total Deaths + Recovered + Active Cases

```python
df['Total_Deaths'] = df['Total_deaths'] + df['New_deaths']
df['Active_Cases'] = df['Active_cases'] + df['New_cases'] 
df.head()
```

```python
df.drop(['New_cases','Total_deaths','New_deaths','Active_cases'],axis=1,inplace=True)
df.head()
```

**Dropped the Old columns. Similarly we're updating the values for Total_cases**

```python
df['Total_Cases'] = df['Recovered'] + df['Total_Deaths'] + df['Active_Cases']
df.drop('Total_cases',axis=1,inplace=True)
df.head()
```

# Acquiring Data for India Cases.

**Getting data from Official Government Website**

```python
url = 'https://www.mohfw.gov.in/'

source = requests.get(url)
print(source)
```

**Checking whether we're getting data from the correct website**

```python
soup = BeautifulSoup(source.content,'html.parser')
soup.title
```

```python
#Getting table details

table = soup.find_all('table')
print(len(table))
```

**Let's check with single content in table**

```python
table[0].find('tr').text.split('\n')
```

**Let's check with single content in first list in table**

```python
t = table[0]
row_one = t.find_all('tr')[1:-1]
print(row_one[0].find_all('td'))
print('\n')
print(row_one[0].find_all('td')[1].text) #Getting state/UT name
```

**Let's do with entire table in to acquire the details**

```python
state_UT = []
confirmed = []
recovered = []
death = []

rows = t.find_all('tr')[1:-1]

for row in rows[:-3]:
    col = row.find_all('td')
    state_UT.append(col[1].text.strip())
    confirmed.append(col[2].text.strip())
    recovered.append(col[3].text.strip())
    death.append(col[4].text.strip())
    
print(state_UT)
print(confirmed)
print(recovered)
print(death)
```

**Construct a dataframe with the scrapped details**

```python
df_india = pd.DataFrame(list(zip(state_UT,recovered,death,confirmed)),columns=['State','Recovered','Total_Deaths',
                                                                              'Total_Cases'])
df_india.head()
```

**Let's verify the datatype before proceeding for data cleaning**

```python
df_india.info()
```

**Convert the Object datatypes to Integer**

```python
cols = ['Total_Cases','Recovered','Total_Deaths']

for col in cols:
    df_india[col] = df_india[col].astype(int)
```

```python
df_india.info()
```

**Generate Active Cases from the existing features**

```python
df_india['Active_Cases'] = df_india['Total_Cases'] - (df_india['Recovered'] + df_india['Total_Deaths'])
df_india.head()
```
